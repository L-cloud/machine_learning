```python
import numpy as np
import matplotlib.pyplot as plt

N = 20
x = np.random.rand(N)
y = 5 + 2*x + 1e-1*np.random.randn(N)
y = y.reshape(N,1)

plt.figure()
plt.scatter(x, y)

# normal equation

# bias 없을 때
X = x.reshape(N,1)
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

y_predict = X*theta
SSE = np.sum((y - y_predict)**2)
print('SSE: %8.3f' % SSE)

x_new = np.linspace(0, 1, 10)
x_new = x_new.reshape(10,1)
y_new = x_new*theta

plt.plot(x_new, y_new)

# bias 있을 때
X = np.hstack([np.ones((N,1)), x.reshape(N,1)])
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

y_predict = X.dot(theta)
SSE = np.sum((y - y_predict)**2)
print('SSE: %8.3f' % SSE)

print(theta)

x_new = np.hstack([np.ones((10,1)), np.linspace(0, 1, 10).reshape(10,1)])
y_new = x_new.dot(theta)

plt.plot(x_new[:,1], y_new)
```

```python
# batch gradient descent
# bias 있을 때

eta = 0.01
maxIter = 20
theta = np.random.randn(2,1)

plt.figure()
plt.scatter(x, y)

y_predict = X.dot(theta)
SSE = [np.sum((y - y_predict)**2)]
print('SSE: %8.3f' % SSE[-1])

y_predict = X.dot(theta)
y_new = x_new.dot(theta)

plt.plot(x_new[:,1], y_new)


for iter in range(maxIter):
    grad = 2*X.T.dot(y_predict - y)
    theta = theta - eta*grad

    y_predict = X.dot(theta)
    SSE.append(np.sum((y - y_predict)**2))
    print('SSE: %8.3f' % SSE[-1])
    
    y_predict = X.dot(theta)
    y_new = x_new.dot(theta)
    
    plt.plot(x_new[:,1], y_new)
    
plt.figure()
plt.plot(SSE)
```

```python
# stochastic gradient descent
# bias 있을 때

eta = 0.1
maxIter = 20
theta = np.random.randn(2)

plt.figure()
plt.scatter(x, y)

y_predict = X.dot(theta)
SSE = [np.sum((y - y_predict)**2)]
print('SSE: %8.3f' % SSE[-1])

y_predict = X.dot(theta)
y_new = x_new.dot(theta)

plt.plot(x_new[:,1], y_new)


for iter in range(maxIter):
    idn = np.random.randint(0, N, 1)
    grad = 2*(y_predict[idn] - y[idn])*X[idn,:].reshape(2,1)
    theta = theta - eta*grad

    y_predict = X.dot(theta)
    SSE.append(np.sum((y - y_predict)**2))
    print('SSE: %8.3f' % SSE[-1])
    
    y_predict = X.dot(theta)
    y_new = x_new.dot(theta)
    
    plt.plot(x_new[:,1], y_new)
    
plt.figure()
plt.plot(SSE)
```

```python
data = np.loadtxt("inclass_06_data.csv", delimiter = ",")

plt.figure()
plt.scatter(data[:,0], np.sum(data[:,1:], axis = 1))

# 출석 / 사전강의&퀴즈 / 과제 / 중간고사 / 기말고사
# pred = 0.10*data[:,1] + 0.15*data[:,2] + 0.20*data[:,3] + 0.25*data[:,4] + 0.30*data[:,5]
```

```python
X = data[:,1:]
N,D = X.shape

y = data[:,0]
y = y.reshape(N,1)

# normal equation
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

y_predict = X.dot(theta)
SSE = np.sum((y - y_predict)**2)
print('SSE: %8.3f' % SSE)

plt.figure()
plt.scatter(y, y_predict)
plt.xlabel('true')
plt.ylabel('predict')

# batch gradient descent
eta = 0.0000001
maxIter = 10000
theta = np.random.randn(D,1)

y_predict = X.dot(theta)
SSE = [np.sum((y - y_predict)**2)]
# print('[%3d] SSE: %8.3f' % (-1, SSE[-1]))

for iter in range(maxIter):
    grad = 2*X.T.dot(y_predict - y)
    theta = theta - eta*grad
    
    y_predict = X.dot(theta)
    SSE.append(np.sum((y - y_predict)**2))
#     print('[%3d] SSE: %8.3f' % (iter, SSE[-1]))
        
plt.figure()
plt.plot(SSE)

plt.figure()
plt.scatter(y, y_predict)
plt.xlabel('true')
plt.ylabel('predict')
```

```python
# stochastic gradient descent
eta = 0.0000001
maxIter = 1000000
theta = np.random.randn(D,1)

y_predict = X.dot(theta)
SSE = [np.sum((y - y_predict)**2)]
# print('[%3d] SSE: %8.3f' % (-1, SSE[-1]))

for iter in range(maxIter):
    idn = np.random.randint(0, N, 1)
    grad = 2*(y_predict[idn] - y[idn])*X[idn,:].reshape(D,1)
    theta = theta - eta*grad
    
    y_predict = X.dot(theta)
    SSE.append(np.sum((y - y_predict)**2))
#     print('[%3d] SSE: %8.3f' % (iter, SSE[-1]))
        
plt.figure()
plt.plot(SSE)

plt.figure()
plt.scatter(y, y_predict)
plt.xlabel('true')
plt.ylabel('predict')
```
